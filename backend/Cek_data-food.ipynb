```python
# =======================================================
# Library Wajib untuk Penelitian:
# 1. Pandas: Untuk memanipulasi data CSV
# 2. os: Untuk cek path file (best practice)
# =======================================================
import pandas as pd
import os
```

```python
# Definisikan path file CSV kamu
# Asumsi: notebook ini dijalankan dari folder 'backend' 
# atau kamu sudah mengupload file CSV ke Colab/Notebook
file_path = 'dataset/MM-Food-100k.csv'

# --- (Bagian 1: Pengecekan dan Pembacaan Data) ---
print(f"Mencoba membaca data dari: {file_path}")

# Cek keberadaan file
if not os.path.exists(file_path):
    print("----------------------------------------------------------")
    print(f"ERROR: File '{file_path}' tidak ditemukan.")
    print("Pastikan file sudah di-upload atau path 'dataset/MM-Food-100k.csv' sudah benar.")
    print("----------------------------------------------------------")
else:
    try:
        # Baca CSV ke dalam DataFrame
        df = pd.read_csv(file_path)

        # Tampilkan ringkasan data (wajib buat penguji)
        print("\n--- 1. Informasi Dasar Data (Shape & Dtypes) ---")
        print(f"Jumlah baris (observasi): {df.shape[0]}")
        print(f"Jumlah kolom (fitur): {df.shape[1]}")
        df.info()

        # --- (Bagian 2: Tampilkan 10 Data Awal dan Akhir) ---

        # Tampilkan 10 data awal
        print("\n--- 2. 10 Data Awal (.head(10)) ---")
        display(df.head(5)) # Pakai 'display' biar outputnya rapi di Notebook

        # Tampilkan 10 data akhir
        print("\n--- 3. 10 Data Akhir (.tail(10)) ---")
        display(df.tail(5)) # Pakai 'display' biar outputnya rapi di Notebook

    except Exception as e:
        print(f"Terjadi error saat membaca/memproses file: {e}")
```

```python
# ==========================================
# Ini Library depencies yang akan digunakan
# di tahap kedua Visualisasi
# ==========================================
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# -------------------------------------------------------------
# Contoh Visualisasi Distribusi Food Type (Kolom Kategorikal)
# -------------------------------------------------------------

# Ambil 10 'food_type' teratas
top_n = 10 
top_food_types = df['food_type'].value_counts().nlargest(top_n)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_food_types.index, y=top_food_types.values, palette='viridis', hue=top_food_types.index, legend=False)
plt.title(f'Top {top_n} Food Types in Dataset')
plt.xlabel('Food Type')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right') # Putar label biar nggak numpuk
plt.tight_layout()
plt.show()
```

```python
# -------------------------------------------------------------
# Contoh Visualisasi Distribusi Food Probability (Kolom Numerik)
# -------------------------------------------------------------

plt.figure(figsize=(8, 5))
sns.histplot(df['food_prob'], bins=20, kde=True)
plt.title('Distribution of Food Probability')
plt.xlabel('Food Probability')
plt.ylabel('Frequency')
plt.show()
```

```python
# ==================================================
# Interaktif Search + Gambar Depencies untuk 
# keperluan pengecekan data penelitian
# ==================================================

from ipywidgets import interactive, Text, Button, VBox, HBox, Output, Layout, Label
from IPython.display import display, HTML
import io
import base64

# Container untuk menampung output gambar dan hasil
output_area = Output(layout=Layout(border='1px solid #ccc', padding='10px'))
search_input = Text(
    value='',
    placeholder='Masukkan keyword (cth: chicken, dessert, salad)',
    description='Keyword:',
    disabled=False,
    layout=Layout(width='90%')
)
search_button = Button(
    description='Cari & Tampilkan Gambar',
    button_style='success', # Warna hijau
    tooltip='Klik untuk mencari dan menampilkan hasilnya',
    layout=Layout(width='200px')
)

# --- FUNGSI UTAMA PENCARIAN & TAMPIL GAMBAR ---
def perform_search(keyword):
    """Mencari dish_name berdasarkan keyword dan menampilkan hasil + gambar."""
    
    # 1. Clear output lama
    output_area.clear_output()

    with output_area:
        if not keyword:
            print("Masukkan keyword yang valid, Lycus.")
            return

        # 2. Filter data (case-insensitive)
        result_df = df[
            df['dish_name'].str.contains(keyword, case=False, na=False)
        ]
        
        if result_df.empty:
            print(f"❌ Tidak ditemukan hidangan dengan keyword '{keyword}'. Coba yang lain!")
            return
        
        # Ambil 5 data teratas untuk ditampilkan
        display_df = result_df.head(5).copy()
        
        print(f"✅ Ditemukan {len(result_df)} hasil untuk '{keyword}'. Menampilkan 5 data teratas dan gambarnya:")

        # 3. Buat HTML untuk menampilkan gambar dan data
        html_content = []
        for index, row in display_df.iterrows():
            # Styling card modern
            card_html = f"""
            <div style="border: 1px solid #ddd; border-radius: 8px; margin-bottom: 20px; padding: 15px; background-color: #f9f9f9; box-shadow: 2px 2px 5px rgba(0,0,0,0.1);">
                <h3 style="color: #333; border-bottom: 2px solid #5cb85c; padding-bottom: 5px;">{row['dish_name']} (ID: {index})</h3>
                <div style="display: flex; gap: 15px;">
                    <div style="flex: 1;">
                        <img src="{row['image_url']}" alt="{row['dish_name']}" style="width: 100%; max-width: 250px; height: auto; border-radius: 4px; object-fit: cover;">
                    </div>
                    <div style="flex: 2; font-size: 14px;">
                        <p><strong>Food Type:</strong> {row['food_type']}</p>
                        <p><strong>Ingredients:</strong> {row['ingredients']}</p>
                        <p><strong>Probabilitas Food:</strong> {row['food_prob']:.2f}</p>
                        <p><strong>Cooking Method:</strong> {row['cooking_method'] if pd.notna(row['cooking_method']) else 'N/A'}</p>
                    </div>
                </div>
            </div>
            """
            html_content.append(card_html)
        
        # Tampilkan semua card dalam satu blok HTML
        display(HTML("".join(html_content)))
        
        # Tampilkan juga tabel ringkas 
        print("\nDetail Tabel 5 Hasil Teratas:")
        display(display_df)

# --- WIDGET HANDLING ---

# Menghubungkan tombol dengan fungsi pencarian
def on_button_clicked(b):
    perform_search(search_input.value)

search_button.on_click(on_button_clicked)

# --- TAMPILKAN UI ---

# Susun UI menjadi HBox (input+tombol) dan VBox (output)
ui = VBox([
    HBox([search_input, search_button], layout=Layout(align_items='center')),
    output_area
])

print("🚀 UI PENCARIAN CANGGIH SUDAH AKTIF 🚀")
display(ui)

# Tampilkan data acak saat pertama kali di-load
with output_area:
    print("Selamat datang! Ketik keyword di atas untuk memulai pencarian interaktif.")
    print("\nMenampilkan 5 sampel acak saat ini:")
    display(df.sample(5))
```
